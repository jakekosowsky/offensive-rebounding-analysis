# Offensive Rebounding Analysis

Explore second-chance opportunities using NBA play-by-play events.


```python
import pandas as pd

pbp = pd.read_csv('../data/2019-20_pbp.csv')
```

## Example: reconstructing a game's opening quarter


```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
df2 = pbp[(pbp['Date'] == 'October 22 2019') & (pbp['AwayTeam'] == 'NOP') & (pbp['Quarter'] == 1)].sort_values(by=['SecLeft'], ascending = False)
df2
#pbp['NumberOfPoints'] = 
```

```python
df2['AwayScore'] = df2['AwayScore'].astype('int')
df2['HomeScore'] = df2['HomeScore'].astype('int')

df2['points'] = (df2['HomeScore'] - df2['HomeScore'].shift(1)) + (df2['AwayScore']- df2['AwayScore'].shift(1)) 
df2.groupby(['Shooter']).agg({'points':['sum','mean'], 'ShotDist': 'mean'})
```
